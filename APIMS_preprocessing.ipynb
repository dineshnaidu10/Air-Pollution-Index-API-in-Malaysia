```python
# Import necessary library and load data into dataframe
import pandas as pd
apims = pd.read_excel('APIMS.xlsx')
```

```python
# Change time to datetime format
apims['Time'] = apims['Time'].dt.strftime('%d/%m/%Y')
apims['Time'] = pd.to_datetime(apims['Time'])
```

```python
# Fill any null rows with 0
apims.fillna(0, inplace=True)
```

```python
# Groupby the datafrime into a single date using mean as the aggregation method
apims2 = apims.groupby(['Time']).mean()
```

```python
# View first 20 rows of the dataframe
apims2.head(20)
```

```python
# Dump the processed file into an excel file
apims2.to_excel('APIMS3.xlsx')
```
